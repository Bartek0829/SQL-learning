```python
%%sql

WITH OrdersByStoreCountry AS (
  SELECT
    st.countryname,
    COUNT(DISTINCT s.orderkey) AS orders
  FROM sales s
  JOIN store st
  USING(storekey)
  GROUP BY st.countryname
)
SELECT
  SUM(orders) AS TotalOrders
FROM OrdersByStoreCountry;
```

```python
%%sql
SELECT
  st.countryname,
  PERCENTILE_CONT(0.50) WITHIN GROUP (ORDER BY(CASE WHEN c.age < 30 THEN (s.netprice*s.quantity*s.exchangerate) END)) AS young,
  PERCENTILE_CONT(0.50) WITHIN GROUP (ORDER BY(CASE WHEN c.age BETWEEN 30 AND 50 THEN (s.netprice*s.quantity*s.exchangerate) END)) AS middle_age,
  PERCENTILE_CONT(0.50) WITHIN GROUP (ORDER BY(CASE WHEN c.age > 50 THEN (s.netprice*s.quantity*s.exchangerate) END)) AS senior
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
LEFT JOIN store st ON s.storekey = st.storekey
GROUP BY st.countryname
```

```python
%%sql
SELECT
  p.categoryname,
  SUM(CASE WHEN c.age < 25 THEN (s.netprice*s.quantity*s.exchangerate) END) AS young,
  SUM(CASE WHEN c.age BETWEEN 25 AND 44 THEN (s.netprice*s.quantity*s.exchangerate) END) AS middle_age,
  SUM(CASE WHEN c.age > 45 THEN (s.netprice*s.quantity*s.exchangerate) END) AS old
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
LEFT JOIN product p ON s.productkey = p.productkey
GROUP BY
  p.categoryname
```

```python
%%sql
SELECT
    p.categoryname,
    CASE
        WHEN c.age < 25 THEN '<25'
        WHEN c.age BETWEEN 25 AND 44 THEN '25-44'
        ELSE '45+'
    END AS age_group,
    SUM(s.quantity * s.netprice * s.exchangerate) AS total_purchases
FROM sales s
INNER JOIN customer c ON s.customerkey = c.customerkey
INNER JOIN product p ON s.productkey = p.productkey
GROUP BY
    age_group,
    p.categoryname;
```

```python
%%sql
SELECT
    s.customerkey,
    CASE
        WHEN SUM(s.quantity * s.netprice * s.exchangerate) < 500 THEN 'Low Spender'
        WHEN SUM(s.quantity * s.netprice * s.exchangerate) BETWEEN 500 AND 2000 THEN 'Medium Spender'
        ELSE 'High Spender'
    END AS spender,
    SUM(s.quantity * s.netprice * s.exchangerate) AS total_purchases
FROM sales s
WHERE s.orderdate BETWEEN '01-01-2023' AND '12-31-2023'
GROUP BY
    s.customerkey
ORDER BY
    s.customerkey
```

```python
%%sql
SELECT
    CASE
      WHEN weight IS NULL OR weightunit IS NULL THEN '1 - No Weight Specified'
      WHEN weightunit = 'pounds' AND weight < 5 THEN '2 - Very Light (< 5 lbs)'
      WHEN weightunit = 'pounds' AND weight BETWEEN 5 AND 25 THEN '3 - Light (5-25 lbs)'
      WHEN weightunit = 'pounds' AND weight BETWEEN 25 AND 100 THEN '4 - Medium (25-100 lbs)'
      WHEN weightunit = 'pounds' AND weight > 100 THEN '5 - Heavy (> 100 lbs)'
      WHEN weightunit = 'ounces' AND weight < 5 THEN '6 - Light Ounces (< 5 oz)'
      WHEN weightunit = 'ounces' AND weight >= 5 THEN '7 - Heavy Ounces (â‰¥ 5 oz)'
      WHEN weightunit = 'grams' THEN '8 - Metric Weights'
      ELSE '9 - Other Weight Categories'
    END AS weight_category,
    COUNT(productkey) AS count_category
FROM
    product
GROUP BY
    weight_category
ORDER BY
    weight_category
```

```python
%%sql
WITH store_revenue AS (
  SELECT
      SUM(s.netprice*s.quantity*s.exchangerate) AS revenue,
      st.storekey AS storekey,
      CASE
        WHEN st.squaremeters < 1000 AND SUM(s.netprice*s.quantity*s.exchangerate) < 100000 THEN '1 - Small Store - Low Revenue'
        WHEN st.squaremeters < 1000 AND SUM(s.netprice*s.quantity*s.exchangerate) >= 100000 THEN '2 - Small Store - High Revenue'
        WHEN st.squaremeters BETWEEN 1000 AND 2000 AND SUM(s.netprice*s.quantity*s.exchangerate) < 300000 THEN '3 - Medium Store - Low Revenue'
        WHEN st.squaremeters BETWEEN 1000 AND 2000 AND SUM(s.netprice*s.quantity*s.exchangerate) >= 300000 THEN '4 - Medium Store - High Revenue'
        WHEN st.squaremeters > 2000 AND SUM(s.netprice*s.quantity*s.exchangerate)< 500000 THEN '5 - Large Store - Low Revenue'
        WHEN st.squaremeters > 2000 AND SUM(s.netprice*s.quantity*s.exchangerate)>= 500000 THEN '6 - Large Store - High Revenue'
        ELSE '7 - Online Store'
      END AS store_category
  FROM sales s
  JOIN store st ON st.storekey = s.storekey
  WHERE s.orderdate BETWEEN '01-01-2023' AND '12-31-2023'
  GROUP BY st.storekey
)

SELECT
 store_category,
 (SUM(revenue) / (SELECT SUM(revenue) FROM store_revenue) *100) AS procentage
FROM store_revenue
GROUP BY store_category
ORDER BY store_category
```
