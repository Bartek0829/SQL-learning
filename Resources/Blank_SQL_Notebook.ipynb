```python
%%sql
SELECT
  COUNT(DISTINCT s.orderkey) AS NumberOfOrders,
  c.countryfull
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
GROUP BY c.countryfull
ORDER BY NumberOfOrders DESC
```

```python
%%sql
SELECT st.countryname, COUNT(DISTINCT s.orderkey) orders
FROM sales s
JOIN store st
USING(storekey)
GROUP BY countryname
ORDER BY orders DESC;
```

```python
%%sql

WITH OrdersByStoreCountry AS (
  SELECT
    st.countryname,
    COUNT(DISTINCT s.orderkey) AS orders
  FROM sales s
  JOIN store st
  USING(storekey)
  GROUP BY st.countryname
)
SELECT
  SUM(orders) AS TotalOrders
FROM OrdersByStoreCountry;
```

```python
%%sql
WITH OrdersByStoreCountry AS (
SELECT
  COUNT(DISTINCT s.orderkey) AS NumberOfOrders,
  c.countryfull
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
GROUP BY c.countryfull
ORDER BY NumberOfOrders DESC)
SELECT
  SUM(NumberOfOrders)
FROM OrdersByStoreCountry
```

```python
%%sql
SELECT
  s.orderdate,
  COUNT(DISTINCT CASE WHEN c.continent = 'Europe' THEN c.customerkey END) AS eu_customer,
  COUNT(DISTINCT CASE WHEN c.continent = 'Australia' THEN c.customerkey END) AS au_customer,
  COUNT(DISTINCT CASE WHEN c.continent = 'North America' THEN c.customerkey END) AS usa_customer,
  COUNT(DISTINCT s.customerkey) AS total_customer
FROM
  sales s
LEFT JOIN customer c ON  s.customerkey = c.customerkey
WHERE
  s.orderdate BETWEEN '01-01-2023' AND '12-31-2023'
GROUP BY
  s.orderdate
ORDER BY
  s.orderdate
```

```python
%%sql
SELECT
  p.categoryname AS category,
  SUM(CASE WHEN s.orderdate BETWEEN '01-01-2022' AND '12-31-2022' THEN (s.quantity*s.netprice*s.exchangerate) ELSE 0 END) AS total_revenue_2022,
  SUM(CASE WHEN s.orderdate BETWEEN '01-01-2023' AND '12-31-2023' THEN (s.quantity*s.netprice*s.exchangerate) ELSE 0 END) AS total_revenue_2023
FROM
  sales s
LEFT JOIN
  product p ON s.productkey = p.productkey
GROUP BY
  p.categoryname
```

```python
%%sql
SELECT
  st.storekey,
  st.description,
  COUNT(DISTINCT CASE WHEN c.gender = 'male' THEN s.orderkey END) AS male,
  COUNT(DISTINCT CASE WHEN c.gender = 'female' THEN s.orderkey END) AS female,
  COUNT(DISTINCT s.orderkey) AS total
FROM
  sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
LEFT JOIN store st ON s.storekey = st.storekey
WHERE
  s.orderdate BETWEEN '01-01-2023' AND '12-31-2023'
GROUP BY
  st.storekey
ORDER BY
  st.storekey
```

```python
%%sql
SELECT
  p.color,
  SUM(CASE WHEN (c.age BETWEEN 18 AND 25) THEN (s.netprice * s.quantity * s.exchangerate) ELSE 0 END) AS age_18_25,
  SUM(CASE WHEN (c.age BETWEEN 26 AND 35) THEN (s.netprice * s.quantity * s.exchangerate) ELSE 0 END) AS age_26_35,
  SUM(CASE WHEN (c.age BETWEEN 36 AND 45) THEN (s.netprice * s.quantity * s.exchangerate) ELSE 0 END) AS age_36_45,
  SUM(CASE WHEN (c.age > 45) THEN (s.netprice * s.quantity * s.exchangerate) ELSE 0 END) AS age_more_than_45
FROM sales s
LEFT JOIN customer c ON s.customerkey = c.customerkey
LEFT JOIN product p ON s.productkey = p.productkey
WHERE s.orderdate BETWEEN '01-01-2023' AND '12-31-2023'
GROUP BY
  p.color
```
